# USING BEAUTIFUL SOUP TO SCRAPE FOOTBALL DATA 

```python
#importing packages
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python

base_url = 'https://understat.com/match/'
match = str(input('Please enter match id :'))
url = base_url + match
```

```python
url
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.content, 'lxml')
scripts = soup.find_all('script')
```

```python
scripts
```

```python
# getting only the shot data
strings = scripts[1].string
```

```python
strings
```

```python
# stripping symbols so we only have json data
ind_start = strings.index("('")+2
ind_end = strings.index("')")
json_data = strings[ind_start:ind_end]
json_data = json_data.encode('utf8').decode('unicode_escape')

#converting strings to json format
data = json.loads(json_data)
```

```python
data
```

```python
#convert data from json to dataframe
#step 1 create a for loop to loop through the json dataset
x = []
y = []
xg = []
team = []
data_home = data['h']
data_away = data['a']

for index in range(len(data_home)):
    for key in data_home[index]:
        if key == 'X':
            x.append(data_home[index][key])
        if key == 'Y':
            y.append(data_home[index][key])
        if key == 'xG':
            xg.append(data_home[index][key])
        if key == 'h_team':
            team.append(data_home[index][key])

for index in range(len(data_away)):
    for key in data_away[index]:
        if key == 'X':
            x.append(data_away[index][key])
        if key == 'Y':
            y.append(data_away[index][key])
        if key == 'xG':
            xg.append(data_away[index][key])
        if key == 'a_team':
            team.append(data_away[index][key])
```

```python
#Now we create the data frame
col_names = ['x', 'y', 'xg', 'team']
df = pd.DataFrame([x,y,xg,team], index=col_names)
```

```python
df
```

```python
#Transpose the table
df = df.T
```
